### Reto
Reto: pide edad y autorización (sí/no). Puede si edad>=16 o (edad>=14 y autorización=True).

```
edad = int(input('Edad: '))
aut = input('Autorización (s/n): ').strip().lower() in ('s','si','sí')
puede = (edad >= 16) or ((edad >= 14) and aut)
print(puede)
```
