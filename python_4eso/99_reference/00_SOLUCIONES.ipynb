### Ejercicio 2
Plantilla while para validar input.

````python
def leer_int():
    while True:
        try:
            return int(input('n: '))
        except ValueError:
            print('no válido')
# leer_int()
````

### Ejercicio 3
Plantilla for range(1,N+1) con acumulador.

````python
N = 5
s = 0
for i in range(1, N+1):
    s += i
print(s)
````

### Ejercicio 4
Plantilla recorrer lista y contar pares.

````python
nums = [1,2,3,4,5,6]
pares = 0
for n in nums:
    if n % 2 == 0:
        pares += 1
print(pares)
````

### Ejercicio 5

````python
d = {'ana':7,'luis':5}
for k,v in d.items():
    print(k, v)
````

### Ejercicio 9
3 asserts para `es_par`.

````python
def es_par(n):
    return n % 2 == 0
assert es_par(0)
assert es_par(2)
assert not es_par(3)
print('OK')
````
