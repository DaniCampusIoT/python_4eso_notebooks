### Reto
Reto: conversor con menú usando funciones: 1) km→m 2) m→cm 3) °C→°F 4) salir.

```
def km_a_m(km: float) -> float:
    return km * 1000
def m_a_cm(m: float) -> float:
    return m * 100
def c_a_f(c: float) -> float:
    return c * 9/5 + 32

def menu():
    while True:
        print('1) km→m  2) m→cm  3) °C→°F  4) salir')
        op = input('Opción: ').strip()
        if op == '1':
            km = float(input('km: '))
            print(km_a_m(km))
        elif op == '2':
            m = float(input('m: '))
            print(m_a_cm(m))
        elif op == '3':
            c = float(input('°C: '))
            print(c_a_f(c))
        elif op == '4':
            break
        else:
            print('Opción inválida')

# Ejecuta manualmente:
# menu()
```

```
def paso(secreto: int, intento: int, intentos: int) -> tuple[bool, int]:
    intentos += 1
    if intento == secreto:
        return True, intentos
    return False, intentos

secreto = 7
intentos = 0
for intento in [3, 7]:
    ok, intentos = paso(secreto, intento, intentos)
    if ok:
        print('acertó en', intentos)
        break
```

### Ejercicio 8
`factorial(n)` (n>=0) + asserts para 0,1,5.

```
def factorial(n: int) -> int:
    if n < 0:
        raise ValueError('n debe ser >= 0')
    res = 1
    for i in range(2, n+1):
        res *= i
    return res
assert factorial(0) == 1
assert factorial(1) == 1
assert factorial(5) == 120
print('OK')
```

### Reto
Reto: implementa `es_primo(n)` y crea asserts para n<2, primos pequeños, no primos y 97.

```
def es_primo(n: int) -> bool:
    if n < 2:
        return False
    for d in range(2, int(n**0.5) + 1):
        if n % d == 0:
            return False
    return True

assert es_primo(-1) is False
assert es_primo(0) is False
assert es_primo(1) is False
assert es_primo(2) is True
assert es_primo(3) is True
assert es_primo(4) is False
assert es_primo(9) is False
assert es_primo(97) is True
print('OK')
```
