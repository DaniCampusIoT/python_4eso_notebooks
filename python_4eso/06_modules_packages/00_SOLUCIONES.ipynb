### Reto
Reto: 'piedra, papel o tijera' contra máquina usando random.

```
import random

def gana(j1: str, j2: str) -> int:
    # devuelve 1 si gana j1, -1 si pierde, 0 empate
    if j1 == j2:
        return 0
    if (j1, j2) in {('piedra','tijera'), ('tijera','papel'), ('papel','piedra')}:
        return 1
    return -1

def juego():
    opciones = ['piedra','papel','tijera']
    while True:
        u = input('Elige piedra/papel/tijera (o salir): ').strip().lower()
        if u == 'salir':
            break
        if u not in opciones:
            print('inválido')
            continue
        m = random.choice(opciones)
        r = gana(u, m)
        print('máquina:', m)
        print('resultado:', 'ganas' if r==1 else 'pierdes' if r==-1 else 'empate')

# Ejecuta manualmente:
# juego()
```
