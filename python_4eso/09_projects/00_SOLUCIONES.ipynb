### Reto
Reto: opción 'usar resultado anterior' como primer operando.

```
def sumar(a,b): return a+b
def restar(a,b): return a-b
def multiplicar(a,b): return a*b
def dividir(a,b): return None if b==0 else a/b

def menu():
    last = 0.0
    while True:
        print('1) sumar 2) restar 3) mult 4) div 5) usar last+sumar 6) salir')
        op = input('Opción: ').strip()
        if op == '6':
            break
        if op in {'1','2','3','4'}:
            a = float(input('a: '))
            b = float(input('b: '))
            if op == '1': last = sumar(a,b)
            if op == '2': last = restar(a,b)
            if op == '3': last = multiplicar(a,b)
            if op == '4':
                r = dividir(a,b)
                if r is None:
                    print('No se puede dividir entre 0')
                    continue
                last = r
            print('=', round(last, 2))
        elif op == '5':
            b = float(input('b: '))
            last = sumar(last, b)
            print('=', round(last, 2))
        else:
            print('Opción inválida')

# Ejecuta manualmente:
# menu()
```

### Reto
Reto: añade niveles (fácil 1..20, normal 1..100, difícil 1..500) y límites según nivel.

```
import random

def config_por_nivel(nivel: str) -> tuple[int,int,int]:
    if nivel == '1':
        return 1, 20, 6
    if nivel == '2':
        return 1, 100, 10
    return 1, 500, 12

def jugar():
    while True:
        nivel = input('Nivel 1=fácil 2=normal 3=difícil: ').strip()
        lo, hi, max_int = config_por_nivel(nivel)
        secreto = random.randint(lo, hi)
        for intento_num in range(1, max_int+1):
            try:
                x = int(input(f'Intento {intento_num}/{max_int}: '))
            except ValueError:
                print('No es número')
                continue
            if x < secreto:
                print('más alto')
            elif x > secreto:
                print('más bajo')
            else:
                print('¡acertaste!')
                break
        else:
            print('Perdiste. Era', secreto)
        if input('¿Otra vez? (s/n): ').strip().lower() not in ('s','si','sí'):
            break

# Ejecuta manualmente:
# jugar()
```

### Ejercicio 7
Función guardar(agenda,fichero) nombre,teléfono por línea.

```
def guardar(agenda, fichero):
    with open(fichero, 'w', encoding='utf-8') as f:
        for nombre, tel in agenda.items():
            f.write(f'{nombre},{tel}\n')
agenda={'ana':'600'}
guardar(agenda,'agenda.txt')
print('ok')
```

### Ejercicio 8
Función cargar(fichero) → dict.

```
def cargar(fichero):
    agenda = {}
    with open(fichero, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            nombre, tel = line.split(',', 1)
            agenda[nombre] = tel
    return agenda
print(cargar('agenda.txt'))
```

### Ejercicio 3
Función `formatea_lcd(temp, hum)` → 2 strings de 14 chars.

```
def formatea_lcd(temp, hum):
    l1 = f'TEMP: {temp:>4.1f}C'
    l2 = f'HUM: {hum:>3d}%'
    return l1[:14].ljust(14), l2[:14].ljust(14)
print(formatea_lcd(24.3, 55))
```

### Ejercicio 4
Imprime marco LCD + 2 líneas usando ljust.

### Ejercicio 10
Modo alerta: si temp>28 o hum>70 imprime ALERTA en 2ª línea.

```
def lcd_con_alerta(temp, hum):
    l1 = f'TEMP: {temp:>4.1f}C'
    if temp > 28 or hum > 70:
        l2 = 'ALERTA'
    else:
        l2 = f'HUM: {hum:>3d}%'
    return l1[:14].ljust(14), l2[:14].ljust(14)
print(lcd_con_alerta(29.0, 50))
```

### Reto
Reto: menú: 1) lectura actual 2) histórico 3) resumen 4) salir.

```
import random

def leer_temperatura():
    return round(random.uniform(18.0, 30.0), 1)
def leer_humedad():
    return random.randint(30, 80)

def lectura_actual():
    return (leer_temperatura(), leer_humedad())

def resumen(lecturas):
    if not lecturas:
        return None
    mt = sum(t for t,_ in lecturas)/len(lecturas)
    mh = sum(h for _,h in lecturas)/len(lecturas)
    mx = max(t for t,_ in lecturas)
    return mt, mh, mx

def menu():
    lecturas = []
    while True:
        op = input('1) actual 2) histórico 3) resumen 4) salir: ').strip()
        if op == '1':
            lec = lectura_actual()
            lecturas.append(lec)
            print('lectura:', lec)
        elif op == '2':
            print(lecturas)
        elif op == '3':
            print(resumen(lecturas))
        elif op == '4':
            break
        else:
            print('inválido')

# Ejecuta manualmente:
# menu()
```
