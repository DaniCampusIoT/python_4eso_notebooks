### Reto
Reto: pide nombre, edad y altura (m). Imprime resumen y calcula IMC suponiendo peso fijo 60 kg.

```
nombre = input('Nombre: ')
edad = int(input('Edad: '))
altura = float(input('Altura (m): '))
peso = 60.0
imc = peso / (altura ** 2)
print(f'Nombre: {nombre}')
print(f'Edad: {edad}')
print(f'Altura: {altura:.2f} m')
print(f'IMC (peso fijo 60 kg): {imc:.1f}')
```
