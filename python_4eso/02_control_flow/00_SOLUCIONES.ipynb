### Reto
Reto: 'selector de tarifa' por edad y estudiante. <12 paga 0, 12–17 paga 2, adulto 5, pero estudiantes adultos 3.

```
def tarifa(edad: int, estudiante: bool) -> int:
    if edad < 12:
        return 0
    if edad <= 17:
        return 2
    return 3 if estudiante else 5

print(tarifa(10, False))
print(tarifa(16, True))
print(tarifa(25, True))
print(tarifa(25, False))
```

### Ejercicio 10
Pide un número y calcula cuántos dígitos tiene usando while (sin len).

```
n = 1234  # en clase: int(input('Número: '))
if n == 0:
    print(1)
else:
    cnt = 0
    x = abs(n)
    while x > 0:
        x //= 10
        cnt += 1
    print(cnt)
```

### Reto
Reto: menú en bucle: 1) sumar 2) restar 3) ver resultado 4) salir. Empieza con resultado=0.

```
def menu_calculadora():
    resultado = 0
    while True:
        print('1) sumar')
        print('2) restar')
        print('3) ver resultado')
        print('4) salir')
        op = input('Elige: ').strip()
        if op == '1':
            n = int(input('Número: '))
            resultado += n
        elif op == '2':
            n = int(input('Número: '))
            resultado -= n
        elif op == '3':
            print('resultado =', resultado)
        elif op == '4':
            break
        else:
            print('Opción inválida')

# Ejecuta manualmente:
# menu_calculadora()
```

### Ejercicio 9
Recorre 1..100 y cuenta cuántos son primos (usa función auxiliar).

```
def es_primo(n: int) -> bool:
    if n < 2:
        return False
    for d in range(2, int(n**0.5) + 1):
        if n % d == 0:
            return False
    return True

cnt = 0
for n in range(1, 101):
    if es_primo(n):
        cnt += 1
print(cnt)
```
