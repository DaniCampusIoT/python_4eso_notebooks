### Reto
Reto: pide notas hasta 'fin', guarda en lista y al final muestra media, max, min y nº de aprobados.

```
def resumen_notas(notas: list[float]) -> dict:
    if not notas:
        return {'media': 0, 'max': None, 'min': None, 'aprobados': 0}
    media = sum(notas) / len(notas)
    max_n = notas[0]
    min_n = notas[0]
    aprobados = 0
    for n in notas:
        if n > max_n:
            max_n = n
        if n < min_n:
            min_n = n
        if n >= 5:
            aprobados += 1
    return {'media': media, 'max': max_n, 'min': min_n, 'aprobados': aprobados}

# Demo (sin input):
notas = [6, 7.5, 4, 9, 5]
print(resumen_notas(notas))
```

### Reto
Reto: mini-agenda CRUD (nombre→teléfono) evitando sobreescritura accidental con confirmación.

```
def add_contact(agenda: dict, nombre: str, tel: str, overwrite: bool = False) -> None:
    nombre = nombre.strip().lower()
    if nombre in agenda and not overwrite:
        raise ValueError('Contacto ya existe (overwrite=False)')
    agenda[nombre] = tel

agenda = {}
add_contact(agenda, 'Ana', '600111222')
try:
    add_contact(agenda, 'Ana', '999999999')
except ValueError as e:
    print('bloqueado:', e)
add_contact(agenda, 'Ana', '999999999', overwrite=True)
print(agenda)
```

### Reto
Reto: de una frase, lista de palabras limpias + longitudes, y palabra más larga.

```
texto = 'Hola, mundo. Esto es una prueba, de comprehensions.'
for ch in ',.':
    texto = texto.replace(ch, '')
palabras = [p.lower() for p in texto.split() if p.strip()]
longs = [len(p) for p in palabras]
max_len = max(longs) if longs else 0
mas_largas = [p for p in palabras if len(p) == max_len]
print('palabras:', palabras)
print('longitudes:', longs)
print('máxima longitud:', max_len)
print('más largas:', mas_largas)
```
