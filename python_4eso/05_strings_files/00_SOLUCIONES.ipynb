### Reto
Reto: diario (añadir entrada / ver todo / salir).

```
def añadir(diario_path='diario.txt'):
    txt = input('Entrada: ')
    with open(diario_path, 'a', encoding='utf-8') as f:
        f.write(txt + '\n---\n')

def ver(diario_path='diario.txt'):
    try:
        with open(diario_path, 'r', encoding='utf-8') as f:
            print(f.read())
    except FileNotFoundError:
        print('(vacío)')

def menu():
    while True:
        op = input('1) añadir 2) ver 3) salir: ').strip()
        if op == '1':
            añadir()
        elif op == '2':
            ver()
        elif op == '3':
            break
        else:
            print('Opción inválida')

# Ejecuta manualmente:
# menu()
```

### Ejercicio 10
Crea `resumen.csv` con (media,max,min,aprobados) y 1 fila.

```
import csv
notas = []
with open('notas.csv', 'r', newline='', encoding='utf-8') as f:
    r = csv.DictReader(f)
    for row in r:
        notas.append(float(row['nota']))
media = sum(notas)/len(notas)
mx = max(notas)
mn = min(notas)
ap = sum(1 for n in notas if n >= 5)
with open('resumen.csv', 'w', newline='', encoding='utf-8') as f:
    w = csv.writer(f)
    w.writerow(['media','max','min','aprobados'])
    w.writerow([media, mx, mn, ap])
print('ok')
```

### Reto
Reto: sistema de notas (añadir, listar, resumen) persistiendo en CSV.

```
import csv
from pathlib import Path

CSV_PATH = Path('notas_sistema.csv')

def cargar() -> list[dict]:
    if not CSV_PATH.exists():
        return []
    with open(CSV_PATH, 'r', newline='', encoding='utf-8') as f:
        return list(csv.DictReader(f))

def guardar(rows: list[dict]) -> None:
    with open(CSV_PATH, 'w', newline='', encoding='utf-8') as f:
        w = csv.DictWriter(f, fieldnames=['nombre','nota'])
        w.writeheader()
        w.writerows(rows)

def añadir(nombre: str, nota: float) -> None:
    rows = cargar()
    rows.append({'nombre': nombre, 'nota': nota})
    guardar(rows)

def resumen() -> dict:
    rows = cargar()
    notas = [float(r['nota']) for r in rows] if rows else []
    if not notas:
        return {'media': 0, 'max': None, 'min': None, 'aprobados': 0}
    return {
        'media': sum(notas)/len(notas),
        'max': max(notas),
        'min': min(notas),
        'aprobados': sum(1 for n in notas if n >= 5)
    }

# Demo (sin menú interactivo):
añadir('Ana', 7)
añadir('Luis', 5)
print(cargar())
print(resumen())
```
