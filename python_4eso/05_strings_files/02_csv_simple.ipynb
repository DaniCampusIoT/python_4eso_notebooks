## Ejemplos previos

### Ejemplo: crear CSV (nombre,nota)

```
import csv
with open('notas.csv', 'w', newline='', encoding='utf-8') as f:
    w = csv.writer(f)
    w.writerow(['nombre', 'nota'])
    w.writerow(['Ana', 7])
    w.writerow(['Luis', 5])
print('CSV creado')
```
