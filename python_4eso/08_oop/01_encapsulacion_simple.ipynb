### Ejemplo: cuenta con saldo

```
class Cuenta:
    def __init__(self, saldo=0):
        self._saldo = saldo
    def ingresar(self, cantidad):
        if cantidad > 0:
            self._saldo += cantidad
    def saldo(self):
        return self._saldo

c = Cuenta(10)
c.ingresar(5)
print(c.saldo())
```
