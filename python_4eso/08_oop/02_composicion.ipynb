### Ejemplo: grupo que contiene alumnos

```
class Alumno:
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = notas
    def media(self):
        return sum(self.notas)/len(self.notas) if self.notas else 0

class Grupo:
    def __init__(self):
        self.alumnos = []
    def añadir(self, alumno):
        self.alumnos.append(alumno)

g = Grupo()
g.añadir(Alumno('Ana', [7,8,9]))
print(g.alumnos[0].media())
```
