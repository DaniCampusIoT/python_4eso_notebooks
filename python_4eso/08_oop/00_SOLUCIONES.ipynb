### Ejercicio 8
Añade `transferir(otra, cantidad)` con validaciones.

```
class Cuenta:
    def __init__(self, saldo=0):
        self._saldo = saldo
    def ingresar(self, cantidad):
        if cantidad > 0:
            self._saldo += cantidad
    def retirar(self, cantidad):
        if 0 < cantidad <= self._saldo:
            self._saldo -= cantidad
    def transferir(self, otra, cantidad):
        if otra is self:
            return
        if 0 < cantidad <= self._saldo:
            self._saldo -= cantidad
            otra._saldo += cantidad

a = Cuenta(10)
b = Cuenta(0)
a.transferir(b, 7)
print(a._saldo, b._saldo)
```

Reto: menú (while) para operar con Cuenta (ingresar/retirar/ver/salir).

```
class Cuenta:
    def __init__(self, saldo=0):
        self._saldo = saldo
    def ingresar(self, c):
        if c > 0:
            self._saldo += c
    def retirar(self, c):
        if 0 < c <= self._saldo:
            self._saldo -= c
    def ver(self):
        print('saldo:', self._saldo)

def menu():
    cuenta = Cuenta(0)
    while True:
        op = input('1) ingresar 2) retirar 3) ver 4) salir: ').strip()
        if op == '1':
            cuenta.ingresar(float(input('cantidad: ')))
        elif op == '2':
            cuenta.retirar(float(input('cantidad: ')))
        elif op == '3':
            cuenta.ver()
        elif op == '4':
            break
        else:
            print('inválido')

# Ejecuta manualmente:
# menu()
```

Añade método `buscar(nombre)` (devuelve alumno o None).

```
class Alumno:
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = list(notas)
class Grupo:
    def __init__(self):
        self.alumnos=[]
    def buscar(self, nombre):
        nombre = nombre.strip().lower()
        for a in self.alumnos:
            if a.nombre.strip().lower() == nombre:
                return a
        return None

g=Grupo(); g.alumnos=[Alumno('Ana',[7])]
print(g.buscar('ana') is not None, g.buscar('marta'))
```

### Reto
Reto: `ranking()` ordenado por media (desc) y mostrarlo.

```
class Alumno:
    def __init__(self, nombre, notas):
        self.nombre = nombre
        self.notas = list(notas)
    def media(self):
        return sum(self.notas)/len(self.notas) if self.notas else 0

class Grupo:
    def __init__(self):
        self.alumnos = []
    def añadir(self, alumno):
        self.alumnos.append(alumno)
    def ranking(self):
        return sorted(self.alumnos, key=lambda a: a.media(), reverse=True)

g = Grupo()
g.añadir(Alumno('Ana',[7,8,9]))
g.añadir(Alumno('Luis',[5,6,5]))
g.añadir(Alumno('Marta',[9,9,8]))
for a in g.ranking():
    print(a.nombre, a.media())
```
